```python
import matplotlib.pyplot as plt
import statistics

nome_aluno = input("Digite o nome do aluno: ")
serie_aluno = input("Digite a série do aluno: ")

disciplinas = ['Português', 'Matemática', 'História', 'Ciências']

medias = []

for disciplina in disciplinas:
    valores = []
    for i in range(4):
        valor = float(input(f"Informe a média do aluno para {disciplina} na prova {i+1}: "))
        valores.append(valor)

    media = statistics.mean(valores)
    medias.append(media)

indice_maior_media = medias.index(max(medias))
indice_menor_media = medias.index(min(medias))

mediana = statistics.median(medias)

plt.bar(disciplinas, medias)

plt.bar(indice_maior_media, max(medias), color='green', label='Maior Média')
plt.bar(indice_menor_media, min(medias), color='red', label='Menor Média')

plt.xlabel('Disciplinas')
plt.ylabel('Médias')
plt.title('Gráfico de Barras - Médias')

plt.legend()
plt.show()

print("Relatório das Médias:")
print(f"Aluno: {nome_aluno}")
print(f"Série: {serie_aluno}")
for i in range(len(disciplinas)):
    print(f"Média em {disciplinas[i]}: {medias[i]}")

print(f"Maior média: {max(medias)}")
print(f"Menor média: {min(medias)}")
print(f"Mediana das médias: {mediana}")
```
